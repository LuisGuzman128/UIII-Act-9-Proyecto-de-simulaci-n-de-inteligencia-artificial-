```python
# 21/mayo/2026

# Programador:
# Lenguaje de programación, experiencia,
# tipo de proyecto y análisis IA.

# ======================================================
# 💻 CBTis 128
# 📚 Elabora Proyectos de Programación Lógica
# 🤖 Especialidad: Inteligencia Artificial
# 👨‍🏫 Docente: Manuel Nava
# 👨‍🎓 Alumno: Axel Guzman NL = 27
# 💻 Proyecto: Lista Programador IA
# ======================================================

# =========================
# 💻 Lista vacía
# =========================
lista_programadores = []

# =========================
# 📌 Menú principal
# =========================
while True:

    print("\n=========== MENÚ PROGRAMADOR 💻 ===========")
    print("1. Agregar programador")
    print("2. Revisar lenguaje")
    print("3. Analizar experiencia")
    print("4. Mostrar programadores")
    print("5. Salir")

    opcion = int(input("Selecciona una opción: "))

    # ==================================================
    # ➕ AGREGAR PROGRAMADOR
    # ==================================================

    if opcion == 1:

        nombre = input("Ingrese nombre del programador: ")
        lenguaje = input("Ingrese lenguaje de programación: ")
        experiencia = int(input("Ingrese años de experiencia: "))
        proyecto = input("Ingrese tipo de proyecto: ")

        lista_programadores.append([
            nombre,
            lenguaje,
            experiencia,
            proyecto
        ])

        print("\n✅ Programador agregado correctamente")

    # ==================================================
    # 🔍 REVISAR LENGUAJE
    # ==================================================

    elif opcion == 2:

        lenguaje = input("Ingrese lenguaje a revisar: ")

        print("\n=========== ANALISIS IA ===========")

        if lenguaje == "python":
            print("🤖 IA recomienda Machine Learning")

        elif lenguaje == "java":
            print("☕ IA recomienda aplicaciones empresariales")

        elif lenguaje == "javascript":
            print("🌐 IA recomienda desarrollo web")

        else:
            print("💻 Lenguaje válido para programación general")

    # ==================================================
    # 📊 ANALIZAR EXPERIENCIA
    # ==================================================

    elif opcion == 3:

        experiencia = int(input("Ingrese años de experiencia: "))

        print("\n=========== ANALISIS IA ===========")

        if experiencia >= 5:
            print("💻 Programador con experiencia avanzada")

        elif experiencia >= 3:
            print("📚 Programador intermedio")

        else:
            print("🧠 Programador principiante")

    # ==================================================
    # 📋 MOSTRAR PROGRAMADORES
    # ==================================================

    elif opcion == 4:

        print("\n======================================================================================================")
        print(f"{'NOMBRE':<25}{'LENGUAJE':<20}{'EXPERIENCIA':<15}{'PROYECTO':<20}")
        print("======================================================================================================")

        i = 0

        while i < len(lista_programadores):

            print(f"{lista_programadores[i][0]:<25}"
                  f"{lista_programadores[i][1]:<20}"
                  f"{lista_programadores[i][2]:<15}"
                  f"{lista_programadores[i][3]:<20}")

            i += 1

        print("======================================================================================================")

    # ==================================================
    # ❌ SALIR
    # ==================================================

    elif opcion == 5:

        print("\n💻 Programa finalizado")
        break

    # ==================================================
    # ⚠️ OPCIÓN INVÁLIDA
    # ==================================================

    else:

        print("❌ Opción no válida")
```
